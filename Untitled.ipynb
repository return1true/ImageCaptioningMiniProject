```python
#Loading Libraries
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import numpy as np
import os
from keras.applications.resnet50 import ResNet50
from keras.optimizers import Adam
from keras.layers import Dense, Flatten, Input, Convolution2D, Dropout, LSTM, TimeDistributed, Embedding, Bidirectional, Activation, RepeatVector, Concatenate
from keras.models import Sequential, Model
from keras.utils import np_utils
from keras.preprocessing import image, sequence
import random
```

```python
#Loading data
images_dir = os.listdir("/Flicker_data/")

images_path = '/Flicker_data/Flicker8k_Dataset/'
captions_path = '/Flicker_data/Flicker8k_Dataset_txt/Flickr8k.token.txt'
train_path = '/Flicker_data/Flicker8k_Dataset_txt/Flickr_8k.trainImages.txt'
val_path = '/Flicker_data/Flicker8k_Dataset_txt/Flickr_8k.devImages.txt'
test_path = '/Flicker_data/Flicker8k_Dataset_txt/Flickr_8k.testImages.txt'

captions = open(captions_path, 'r').read().split("\n")
x_train = open(train_path, 'r').read().split("\n")
x_val = open(val_path, 'r').read().split("\n")
x_test = open(test_path, 'r').read().split("\n")
print("Data has been loaded!!")
```

```python
# created a dictionary named “Image_description” which contains the name of the image(Image_id) as keys and a list of the 5 captions for the corresponding image as values.
Image_description = {}

for i in range(len(captions)-1):
    temp = captions[i].split("#")
    if temp[0] in Image_description:
        Image_description[temp[0]].append(temp[1][2:])
    else:
        Image_description[temp[0]] = [temp[1][2:]]
print("for example Image_description will look like- \n")
print(Image_description['101654506_8eb26cfb60.jpg'])
```

```python
# Creating training, testing and validation dataset files with header as 'image_id' and 'captions'
train = open('flickr_8k_train_dataset.txt','wb')
train.write(b"image_id\tcaptions\n")

validate = open('flickr_8k_val_dataset.txt','wb')
validate.write(b"image_id\tcaptions\n")

test = open('flickr_8k_test_dataset.txt','wb')
test.write(b"image_id\tcaptions\n")
```

```python
#Populating the above created files for training, testing and validation dataset with image ids and captions for each of these images and
#adding ‘startseq’ and ‘endseq’ with each captions.
#‘startseq’ -> This is a start sequence token which will be added at the start of every caption.
#‘endseq’ -> This is an end sequence token which will be added at the end of every caption.

for img in x_train:
    if img == '':
        continue
    for capt in Image_description[img]:
        caption = "<startseq> "+ capt + " <endseq>"
        train.write((img+"\t"+caption+"\n").encode())
        train.flush()
train.close()

for img in x_test:
    if img == '':
        continue
    for capt in Image_description[img]:
        caption = "<startseq> "+ capt + " <endseq>"
        test.write((img+"\t"+caption+"\n").encode())
        test.flush()
test.close()

for img in x_val:
    if img == '':
        continue
    for capt in Image_description[img]:
        caption = "<startseq> "+ capt + " <endseq>"
        validate.write((img+"\t"+caption+"\n").encode())
        validate.flush()
validate.close()
print("training, testing and validation dataset has been populated ")
```

```python
# Loading 50 layer Residual Network Model and getting the summary of the model

model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3),pooling='avg')
model.summary()
```

```python
from keras.preprocessing import image

# Helper function to process images
def preprocessing(img_path):
    im = image.load_img(img_path, target_size=(224,224,3))
    im = image.img_to_array(im)
    im = np.expand_dims(im, axis=0)
    return im


train_data = {}
count=0
for i in x_train:
    if i == "":
        continue
    if count >= 3000:
        break
    count+=1
    if count%1000==0:
        print(count)
    path = images_path + i
    img = preprocessing(path)
    pred = model.predict(img).reshape(2048)
    train_data[i] = pred
    print(pred)
```

```python
# storing all the cpations into a list
dataset = pd.read_csv("flickr_8k_train_dataset.txt",delimiter='\t')
ds=dataset.values
print(ds.shape)

sentences=[]
for i in range(ds.shape[0]):
    sentences.append(ds[i, 1])
    
print(len(sentences))

dataset.head()
```

```python
#first 10 captions
sentences[:10]
```

```python
# Splitting each captions stored in 'sentences' and storing them in 'words' as list of list
words = [i.split() for i in sentences]

# Creating a list of all unique words
unique = []
for i in words:
    unique.extend(i)
    
unique = list(set(unique))

print(len(unique))

vocab_size = len(unique)

unique2=[]
for w in unique:
    count=0;
    for i in sentences:
        if w in i:
            count=count+1;
            if(count>=10):
                unique2.append(w)
                break;
            
print(len(unique2))
```

```python
# Vectorization
print(unique2[:10])
word2indices = {val:index for index, val in enumerate(unique2)}
indices2word = {index:val for index, val in enumerate(unique2)}

#print(word2indices)

word2indices['UNK'] = 0
word2indices['raining'] = 2288
indices2word[0] = 'UNK'
indices2word[2288] = 'raining'

print('Done!')
```

```python
print(word2indices['<startseq>'])
print(indices2word[2280])
print(word2indices['<endseq>'])
print(indices2word[0])
```

```python
vocab_size = len(word2indices.keys())
print(vocab_size)
```

```python
max_len = 0

for i in sentences:
    i = i.split()
    if len(i) > max_len:
        max_len = len(i)

print(max_len)
```

```python
padded_sequences, subsequent_words = [], []

for ix in range(ds.shape[0]):
    partial_seqs = []
    next_words = []
    text = ds[ix, 1].split(' ')
    #print(text)
    text2=[]
    for j in text:
        if j in unique2:
            text2.append(word2indices[j])
            
    for i in range(1, len(text2)):
        partial_seqs.append(text2[:i])
        next_words.append(text2[i])
    padded_partial_seqs = sequence.pad_sequences(partial_seqs, max_len, padding='post')

    next_words_1hot = np.zeros([len(next_words), vocab_size], dtype=np.bool)
    
    #Vectorization
    for i,next_word in enumerate(next_words):
        next_words_1hot[i, next_word] = 1
        
    padded_sequences.append(padded_partial_seqs)
    subsequent_words.append(next_words_1hot)
    
    
padded_sequences = np.asarray(padded_sequences)
subsequent_words = np.asarray(subsequent_words)

print(padded_sequences.shape)
print(subsequent_words.shape)
```

```python
embedding_size = 128
max_len = 40
```

```python
image_model = Sequential()

image_model.add(Dense(embedding_size, input_shape=(2048,), activation='relu'))
image_model.add(RepeatVector(max_len))

image_model.summary()
```

```python
language_model = Sequential()

language_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_size, input_length=max_len))
language_model.add(LSTM(256, return_sequences=True))
language_model.add(TimeDistributed(Dense(embedding_size)))

language_model.summary()
```

```python
conca = Concatenate()([image_model.output, language_model.output])
x = LSTM(128, return_sequences=True)(conca)
x = LSTM(512, return_sequences=False)(x)
x = Dense(vocab_size)(x)
out = Activation('softmax')(x)
model = Model(inputs=[image_model.input, language_model.input], outputs = out)

# model.load_weights("../input/model_weights.h5")
model.compile(loss='categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
model.summary()
```

```python
# fitting the model with epoch = 200
hist = model.fit([images, captions], next_words, batch_size=512, epochs=200)
```
